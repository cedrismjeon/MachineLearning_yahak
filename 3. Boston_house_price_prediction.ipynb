```
import tensorflow as tf
import pandas as pd
```

```
#파일 불러오기
url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
file = pd.read_csv(url)
print(file)
```

```
#열 이름 불러오기
file.columns
```

```
#독립변수, 종속변수 구분하기
independent = file[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
dependent = file[['medv']]
print(independent.shape, dependent.shape)
```

```
#모델 구조만들기
X = tf.keras.layers.Input(shape=[13]) #독립변수 열이 13개
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
#데이터 fit 하기
model.fit(independent, dependent, epochs=2000, verbose=1)
```

```
#모델을 이용한다
print(model.predict(independent[0:5]))
```

```
# 모델의 수식 확인
print(model.get_weights())
```
